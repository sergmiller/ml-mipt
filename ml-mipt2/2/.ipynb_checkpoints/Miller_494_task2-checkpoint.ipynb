    где $y_t$ стационарный процесс с нулевым средним,$~~$
    $\phi_1,\dots,\phi_p$ - константы ($\phi_p \neq 0$), $~~$

  где $\hat y_t$ - прогноз $y_t$ согласно следующей регрессионной модели $$y_t = \sum_{j=1}^{\tau-1}\beta_j y_{t+j}$$
    $\hat y_{t+\tau}$ - forecast of $y_{t\tau}$ with following regression model $$y_{t+\tau} = \sum_{j=1}^{\tau-1}\gamma_j y_{t+j}.$$

   
   $\mathsf{E}r(e_t, e_{t+\tau}) \propto \mathsf{E} \sum_{i=1}^{p} (e_i - \overline{e}) (e_{i+\tau} - \overline{e})$, где $\overline{e} = \frac{1}{p + \tau}\sum_{t=1}^{p + \tau} e_t. \mathsf{E} \overline{e} = \mathsf{E}e_1 = 0$. При $\tau > p$, $e_i$ не зависит от $e_{i+\tau}$. $\sum_{i=1}^{p} \mathsf{E} (e_i - \overline{e}) \mathsf{E}(e_{i+\tau} - \overline{e}) = 0 \cdot 0 = 0$

```python
def fix_ts(t):
    t = t.dropna(axis=0)
    t = t.apply(lambda s: float(s.replace('..','.').replace(',', '.')) if isinstance(s, str) else s)
    return t

train_ts = fix_ts(train_ts)
```

Визуально дисперсия выглядит не постоянной, поэтому нужно применить преобразование Бокса-Кокса.

```python
np.min(train_ts)
```

```python
train_ts += 1
```

```python
from copy import deepcopy

def get_boxcox(ts, lmbda=None):
    t_boxcox = pd.DataFrame(deepcopy(ts))
    normed = sps.boxcox(ts, lmbda)
    if lmbda is None:
        t_boxcox['ts_boxcox'] = normed[0]
    else:
        t_boxcox['ts_boxcox'] = normed
    t_boxcox = t_boxcox['ts_boxcox']
    return t_boxcox, normed[1]
```

```python
train_ts_boxcox, lmbda = get_boxcox(train_ts)
```

Для выбора d будем дифференцировать исходный ряд, пока не перестанет отвергаться гипотеза о стационарности(сделаем это с помощью множественной проверки гипотез)

```python
from statsmodels.sandbox.stats.multicomp import multipletests  # Holm method
from statsmodels.tsa import stattools
```

```python
p_vals = []
diff = np.array(train_ts_boxcox)
for d in np.arange(0,6):
    print(stattools.kpss(diff))
    p_vals.append(stattools.kpss(diff)[1])
    diff = np.diff(diff)
    
reject = multipletests(pvals=p_vals, method='holm',alpha=0.1)[0]
print(reject, p_vals)
```

По тестовой статистике видно, что для d=0, гипотеза стационарности все же отвергается (хотя тест выдает p_val=0.01, это искуственное ограничение).
Поэтому выберем d=1.

```python
diff_train_ts_boxcox = (train_ts_boxcox  - train_ts_boxcox.shift(1))[1:]
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(diff_train_ts_boxcox, lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(diff_train_ts_boxcox, lags=40, ax=ax2)
```

Так как S=7, то кандидаты для q- 0,1,5,6; для p - 0,1,2,3,4,5,6

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from tqdm import tqdm
best_aic = np.inf
best_pq = None
best_params = None
resid = None
d = 1
for p in tqdm([0,1,2,3,4,5,6], position=0):
    for q in [0,1,5,6]:
        try:
            arma_model=sm.tsa.ARIMA(train_ts_boxcox, (p,d,q)).fit()
            if arma_model.aic < best_aic:
                best_aic = arma_model.aic
                best_pq = (p,q)
                best_params = arma_model.params
                resid = arma_model.resid
        except:
            pass
        
print('best params: \n {}'.format(best_params))
print('p = {}, q = {}; best aic: {}'.format(best_pq[0], 
                                                best_pq[1],
                                                best_aic,
                                                ))
```

Проверим остатки на несмещенность, стационарность и отсутствие автокорелляции множественной проверкой гипотез

```python
tests = [
    lambda r: sps.normaltest(r)[1],
    # estimation of p-value (durbin-watson return statistic btwn 0 and 4)
    lambda r: min(sm.stats.durbin_watson(r), 4 - sm.stats.durbin_watson(r))/2,
    lambda r: sps.wilcoxon(r)[1],
    lambda r: stattools.kpss(r)[1], 
    # dickey-fuller test check for unit root (and non stationarity)
    lambda r: stattools.adfuller(r)[1],
    # student ttest for check mean of acf is equal 0
    lambda r: sps.ttest_1samp(stattools.acf(r), popmean=0)[1]
]


p_vals = []

for t in tests:
    p_vals.append(t(resid))
    
reject = multipletests(pvals=p_vals, method='holm',alpha=0.05)[0]
print(reject, p_vals)
```

Тест Дикки-Фуллера отвергся, как и должно было быть для стационарного ряда.
normaltest отвергся, но это лишь значит что остатки не распределены нормально.

```python
def qualityMACAPE(x,y):
    # Mean average corrected absolute percentage error
    # x,y - pandas structures
    # x - real values
    # y - forecasts
    qlt = (2*np.abs(x-y)/(x+y))
    return qlt.mean() , np.abs(x-y)
```
